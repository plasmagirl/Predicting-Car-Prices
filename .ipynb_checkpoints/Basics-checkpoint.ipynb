Automobile dataset downloaded from the UCI Machine learning repository.

This data set consists of three types of entities: (a) the specification of an auto in terms of various characteristics (performance characterisitics and physical characteristics), (b) its assigned insurance risk rating (symboling), (c) its normalized losses in use as compared to other cars. The second rating corresponds to the degree to which the auto is more risky than its price indicates. Cars are initially assigned a risk factor symbol associated with its price. Then, if it is more risky (or less), this symbol is adjusted by moving it up (or down) the scale. Actuarians call this process "symboling". A value of +3 indicates that the auto is risky, -3 that it is probably pretty safe. 

The third factor is the relative average loss payment per insured vehicle year. This value is normalized for all autos within a particular size classification (two-door small, station wagons, sports/speciality, etc...), and represents the average loss per car per year. 

Can we predict the price of a car given all these characterisitics and with what accuracy?


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from IPython.display import display
pd.set_option('display.max_columns', 30)
cols = ['symboling', 'normalized_losses', 'make', 'fuel_type', 'aspiration', 'num_doors', 'body_style', 'drive_wheels', 'engine_location', 'wheel_base', 'length', 'width', 'height', 'curb_weight', 'engine_type', 'num_cylinders', 'engine_size', 'fuel_system', 'bore', 'stroke', 'compression_ratio', 'horsepower', 'peak_rpm', 'city_mpg', 'highway_mpg', 'price']
cars = pd.read_csv('imports-85.data', header=None, names=cols)
display(cars.sample(5))
display(cars.shape)
```

```python
display(cars.info())
```

The dataframe car consists of 205 rows/instances and 26 attributes. A quick look at the data tells us that the columns are a mix of categorical and quantitative variables. We'll divide the attributes between the two types for easier analysis. Symboling is a categorical(ordinal) variable even though is an int data type. And columns like normalized_losses, bore, stroke, horsepower and price are quantitative variables even though they are categorized as object.

```python
#Seperate the integer columns from object columns
columns_cat = cars.select_dtypes(include='object').columns
columns_cat = columns_cat.drop(['normalized_losses', 'bore', 'stroke', 'horsepower', 'peak_rpm', 'price'])
columns_cat = columns_cat.insert(0, 'symboling')
columns_int = [col for col in cars.columns if col not in columns_cat]
print('Categorical columns are-\n{}' .format(columns_cat))
print('Integer columns are-\n{}' .format(columns_int))
```

The quantitative columns are a mix of object and integer datatype. We are going to convert them all to ints or floats. First lets check for missing values (nan/?)

```python
display(cars.isna().sum()) #check for missing values in all columns
#No nan values found. Check for ? - if any will be in columns with object dtype (int or float won't support '?')
print('True values below indicates the number of missing data(?) in the columns-\n')
for col in cars.select_dtypes(include = 'object').columns:
    missing_vals = (cars[col] == '?')
    print(missing_vals.value_counts())
    
```

Normalized losses have the most unfilled values(41). Columns num_doors, bore, stroke, horsepower and price have 2 to 4 missing values. We're going to replace them with nan values and convert the relevant columns to float/int.

```python
cols = ['normalized_losses', 'num_doors', 'bore', 'stroke', 'horsepower', 'peak_rpm', 'price'] #columns with ? values
for col in cols:
    cars[col].replace(to_replace = '?', value = np.nan, inplace=True)

sns.heatmap(cars.isnull(), cmap='Blues') #heatmap shows that bore, stroke and horsepower, peak_rpm are missing values in the same rows.
display(cars[cols].isna().sum()) #to make sure all ? values have been replaced by np.nan

for col in columns_int:
    if cars[col].dtype == 'object':
        cars[col] = cars[col].astype('float')
```

Normalized losses column has the most missing values. I will drop the column for now. The missing values in the rest of the columns will be imputed as follows - 

num_doors - both rows with missing door data are sedans (4 doors). Replace data with 4 

bore & stroke - bore is the diameter of each cylinder and stroke is defined as the full travel of piston along the cylinder. Will replace them with the mean of the column

horsepower & peak_rpm - replace it with the mean of its body style (wagon and hatchback)

price - mean of the column

```python
cars['num_doors'] = cars['num_doors'].replace(to_replace = np.nan, value = 'four')
cars['bore'] = cars['bore'].replace(to_replace = np.nan, value = cars['bore'].mean())
cars['stroke'] = cars['stroke'].replace(to_replace = np.nan, value = cars['stroke'].mean())
cars['price'] = cars['price'].replace(to_replace = np.nan, value = cars['price'].mean())
```

```python
#return the index of the missing value in the columns which is same for horsepower and peak_rpm as we can see in the heatmap
indexes = cars[cars['horsepower'].isna()==True].index 
display(indexes)
for col in ['horsepower', 'peak_rpm']:
    print(col)
    for idx in indexes:
        #print(idx)
        val = (int(cars[cars['body_style'] == cars.iloc[idx]['body_style']][col].mean())) #calculates the mean of the column with the same body style as the missing row
        #print(val)
        cars.loc[idx, col] = val #this method does not raise a warning unlike cars.iloc[idx][col] = val statement will
#display(cars.iloc[130])
#display(cars.iloc[131])
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
#splitting the dataframe into train and test samples

def train_test(df, frac=0.7):
    train = df.sample(frac=frac, replace=False, random_state=1)
    #display(train.head(5))
    test_idx = [x for x in df.index if x not in train.index]
    test = df.loc[test_idx]
    #display(test.head(5))
    return train, test



train, test = train_test(cars)
```

The heatmap tells us that all the remaining columns are strongly correlated to price. Similarly curb weight is also correlated to almost all the columns. Maybe just the curb_weight might be enough for analysis. But I am going to use each column seperately for regression analysis and calculate the error. If that doesn't work then combine each column with curb_weight(or any other column that gives a high R2 value) to see if there is an improvement in the model. Only after that will I play with categorical columns.

```python
#function takes the columns and returns coeffs, rmse and R2 values

def lr_train_test(feature_X, target_y, train, test):
    lr = LinearRegression()
    lr.fit(train[feature_X], train[target_y])
    predict = lr.predict(test[feature_X])
    rmse = np.sqrt(mean_squared_error(test[target_y], predict))
    r2 = r2_score(test[target_y], predict)
    adj_r2 = 1 - (((1-r2)*(len(test)-1))/(len(test)-len(lr.coef_)-1)) #adjusted R2 metric 
    print('Coefficient for {}: {}' .format(feature_X, lr.coef_))
    return rmse, r2, adj_r2
```

```python
metric_lr = {}
for col in corr_columns:
    metric = lr_train_test([col], 'price', train, test)
    metric_lr[col] = metric

print("RMSE and R2 values of the linear regression model with just one feature -\n")
display(metric_lr)

```

As expected, LR model using curb weight (has one of the highest correlation and is also correlated to other features) as the parameter provides the best results. Next I will add one more feature to the curb weight and check the error metrics.

```python
metric_lr_2feature = {}
columns = corr_columns.drop('curb_weight')
for col in columns:
    metric = lr_train_test(['curb_weight', col], 'price', train, test)
    metric_lr_2feature['curb_weight, '+col] = metric
    
display(metric_lr_2feature)
```

From the first glance, R2 seems to increase with addition of most of the features but adjusted R2(that takes into account the loss in degress of freedom due to the increase in the number of features, hence a better indicator for multiple regression model) has decreased wrt to just the curb weight as the predictor. RMSE is slightly reduced and adjusted R2 is increased with the addition of engine size and length to curb weight but they are also highly collinear to each other (check the heatmap graph). The improvement might just be superficial

```python
display(lr_train_test(['curb_weight', 'engine_size', 'length'], 'price', train, test))
#no improvement in model
```
